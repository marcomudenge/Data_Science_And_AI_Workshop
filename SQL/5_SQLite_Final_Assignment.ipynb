```python
import pandas

schools_df = pandas.read_csv('Chicago_Public_Schools_-_Progress_Report_Cards__2011-2012_.csv')
schools_rows = schools_df.to_sql("SCHOOLS", conn, if_exists='replace', index=False)

census_df = pandas.read_csv('Census_Data_-_Selected_socioeconomic_indicators_in_Chicago__2008___2012.csv')
census_rows = census_df.to_sql("CENSUS", con=con, if_exists='replace', index=False)

crimes_df = pandas.read_csv('Crimes_-_2001_to_Present.csv')
crimes_rows = crimes_df.to_sql("CRIMES", con=con, if_exists='replace', index=False)

print('Number of rows affected to all tables... SCHOOLS: {}, CENSUS: {}, CRIMES: {}'.format(schools_rows, census_rows, crimes_rows))
```

## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
# Data as of 2023-08-14 may differ from the original answer
# Count all crimes recorded in the CRIMES table

%sql SELECT COUNT(*) FROM CRIMES;
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
%sql SELECT `Community Area Name`, `COMMUNITY AREA NAME`, `PER CAPITA INCOME ` FROM CENSUS WHERE `PER CAPITA INCOME ` < 11000;
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
%sql SELECT `Case Number`, `Primary Type`, `Description` FROM CRIMES WHERE `Description` LIKE '%MINOR%' LIMIT 10
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
# Showing only 10 rows
# According to the Chicago Police Department's website, IUCR 1792 refers to the kidnapping of a child

%sql SELECT DISTINCT `Primary Type`, `Description` FROM CRIMES WHERE `Primary Type` = 'KIDNAPPING' AND `Description` LIKE '%CHILD%';
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
%sql SELECT DISTINCT `Primary Type` FROM CRIMES WHERE `Location Description` LIKE '%SCHOOL%';
```

### Problem 6

##### List the average safety score for each type of school.


```python
# Assuming the type refers to elementary, middle or high school
%sql SELECT `Elementary, Middle, or High School` AS 'Type', AVG(`Safety Score`) AS 'Average safety score' \
     FROM SCHOOLS GROUP BY `Elementary, Middle, or High School`;
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
%sql SELECT `Community Area Number`, `COMMUNITY AREA NAME`, `PERCENT HOUSEHOLDS BELOW POVERTY` \
     from CENSUS order by `PERCENT HOUSEHOLDS BELOW POVERTY` DESC limit 5;
```

### Problem 8

##### Which community area is most crime prone?


```python
%sql SELECT `Community Area`, `COMMUNITY AREA NAME`, COUNT(ID) AS 'Crimes' \
     FROM CRIMES, CENSUS \
     ON (`Community Area` = `Community Area Number`) \
     GROUP BY `Community Area` ORDER BY `Crimes` DESC LIMIT 1;
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
%sql SELECT `COMMUNITY AREA NAME` FROM CENSUS \
    WHERE `HARDSHIP INDEX` = (SELECT MAX(`HARDSHIP INDEX`) FROM CENSUS);
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
# various ways to do ths 
%sql SELECT `COMMUNITY AREA NAME`  \
     FROM CENSUS \
     WHERE `Community Area Number` IN (SELECT `Community Area` FROM CRIMES \
                                       WHERE `Community Area` IS NOT NULL \
                                       GROUP BY `Community Area` \
                                       ORDER BY COUNT(`ID`) DESC LIMIT 1);
```
